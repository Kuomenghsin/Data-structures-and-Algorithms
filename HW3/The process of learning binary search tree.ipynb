## BST原理

二元搜尋樹，是資料庫系統的原型，結構必定符合兩點特性：
   - 每個node最多只能有兩個child
   - 新增node不可指定位置，其必定依照與root(parent)相比，大於放右邊，小於、等於放左邊

適用於資料愈亂愈好，若放入資料為整理過後，會變得像linked list一樣，無法無法呈現BST結構搜尋效率的優點

```python
def _delete_pointer(root, target):
    if root == None:
        return
    
    pointer = root
    pre_pointer = pointer
    i = 0
    delete_node = pointer
    pre_delete_node = pre_pointer

    while pointer.left != None or pointer.right != None:
        if target == pointer.val:
            delete_node = pointer
            pre_delete_node = pre_pointer
            i += 1
            pre_pointer = pointer
            pointer = pointer.left

        elif target < pointer.val and pointer.left != None:
            pre_pointer = pointer
            pointer = pointer.left

        elif target > pointer.val and pointer.right != None:
            pre_pointer = pointer
            pointer = pointer.right
        else:
            break
        
    if target == pointer.val:
        delete_node = pointer
        pre_delete_node = pre_pointer
        i += 1
  
    return pre_delete_node, delete_node, i
```

```python
def _delete_root(root, target):
    
    Max = root.left
    pre_Max = root
    
    while Max.right != None:
        pre_Max = Max
        Max = Max.right
        
    if Max == root.left:
        root.val = Max.val
        root.left = Max.left
        
    else:
        if Max.left == None:
            root.val = Max.val
            pre_Max.right = None
        else:
            root.val = Max.val
            pre_Max.right = Max.left     
    
    return root
```

```python
def insert(root, val):
    if root == None:
        return

    if val != None:
        if val <= root.val:
            if root.left == None:
                root.left = TreeNode(val)                
            else:
                insert(root.left, val)                    
        else:
            if root.right == None:
                root.right = TreeNode(val)                
            else:
                insert(root.right, val)
    
    return TreeNode(val)
```
